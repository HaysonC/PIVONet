# Test Simulation
Run a quick particle simulation in a Poiseuille flow and visualize a few trajectories.

```python
import numpy as np
from thermal_flow_cnf.src.config import CONFIG
from thermal_flow_cnf.src.flows import poiseuille_flow
from thermal_flow_cnf.src.simulation.langevin import simulate_dataset
from thermal_flow_cnf.src.evaluation.visualize import plot_trajectories

cfg = CONFIG.copy()
cfg["T"] = 500
cfg["num_particles"] = 200
flow_fn = poiseuille_flow(cfg["Umax"], cfg["H"])
out = simulate_dataset(
    flow_fn,
    cfg["num_particles"],
    cfg["D"],
    cfg["dt"],
    cfg["T"],
    cfg["H"],
    save_dir="thermal_flow_cnf/data/raw",
    prefix="notebook",
)
data = np.load(out)
plot_trajectories(data["trajs"], H=cfg["H"], n_show=50)
```
